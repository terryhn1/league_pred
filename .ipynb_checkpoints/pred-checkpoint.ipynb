```python
import pyGMs as gm
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
import data_extraction
import factors

%matplotlib inline
```

```python
"""
    Variables can use an average of training data to determine the states seen.

    Jungle Monsters Killed: 16 States(4-bit number indicating red/blue teamplay indicator + team lead diff)

    Turrets Destroyed: 4 States(max of 3 turrets can be destroyed early game if large teamplay)

    Total CS Difference: 2 States(average threshold)

    Ward Score Difference: 3 States(0 for no team advantage, 1 for blue team advantage, 2 for red team advantage)

    Elite Monsters: 3 States(max of 2 elite monsters can be taken within first 10 minutes practically)

    Gold Difference: 2 States(average threshold)

    Experience Difference: 2 states(average threshold)

    Teamplay: 4 States(Based off a Teamplay Score Rating)

    Lane Dominance: 4 States(Based off a Lane Dominance Rating)

    KDA: 2 States(Average KDA)

    Win Condition: 2 states(Win or Lose)
"""
def initializeGraphModel():

    #Initialize the Variables
    jgMonstersKilled = gm.Var(0, 16)
    turretsDestroyed = gm.Var(1, 3)
    csDifference = gm.Var(2, 3)
    wardScoreDifference = gm.Var(3, 3)
    eliteMonsters = gm.Var(4, 3)
    goldDifference = gm.Var(5, 3)
    expDifference = gm.Var(6, 3)
    avgKDA = gm.Var(7, 3)
    laneDominance = gm.Var(8, 3)
    teamplayScore = gm.Var(9, 3)
    winCondition = gm.Var(10, 2)
    
    Vars = [jgMonstersKilled, turretsDestroyed, csDifference, wardScoreDifference,
            eliteMonsters, goldDifference, expDifference, avgKDA, laneDominance,
            teamplayScore, winCondition]

    # Initialize the factors with Dirichlet prior 
    
    f1 = gm.Factor([jgMonstersKilled], 1.0)
    f2 = gm.Factor([turretsDestroyed], 1.0)
    f3 = gm.Factor([csDifference], 1.0)
    f4 = gm.Factor([wardScoreDifference], 1.0)
    f5 = gm.Factor([eliteMonsters], 1.0)
    f6 = gm.Factor([goldDifference, jgMonstersKilled, csDifference, eliteMonsters], 1.0)
    f7 = gm.Factor([expDifference, jgMonstersKilled, csDifference], 1.0)
    f8 = gm.Factor([avgKDA], 1.0)
    f9 = gm.Factor([laneDominance, goldDifference, expDifference, avgKDA], 1.0)
    f10 = gm.Factor([teamplayScore, jgMonstersKilled, turretsDestroyed, wardScoreDifference, eliteMonsters], 1.0)
    f11 = gm.Factor([winCondition, laneDominance, teamplayScore], 1.0)

    return gm.GraphModel([f1,f2, f3, f4, f5, f6, f7, f8, f9, f10, f11]), Vars
```

```python
def loadData():
    data = np.genfromtxt("data.csv", delimiter = ",")
    
    data_int = np.array([list(xj) for xj in data], dtype=int)-1
    
    nTrain = int(.75*len(data_int))
    train = data_int[:nTrain]
    valid = data_int[nTrain:]
    
    return train, valid
```

```python
'''
Uses train data and the Chow-Liu algorithm to build the 
maximum likelihood tree model structure and returns phat, 
the probability estimates for single-variable and pairwise
probabilities. phat will be used later on to build a better
estimated model.
'''
def estimateProb(model, train):
    phat = {}
    
    # estiamte single-variable probabilities
    for i in model.X:
        phat[i] = gm.Factor([i],1e-15)
        for xs in train:
            phat[i][xs[i]] += 1.0
        phat[i] /= len(train)
      
    # estimate pairwise probabilities
    for i in model.X:        
        for j in model.X:
            if j<=i: continue
            phat[i,j] = gm.Factor([i,j],1e-15)
            for xs in train: 
                phat[i,j][xs[i],xs[j]] += 1.0
            phat[i,j] /= len(train)

    return phat
```

```python
'''
Compute empirical mutual information and show the
correlation between features/variables
'''
def computeEmpirical(model, phat): 
    wts = np.zeros((train.shape[1],train.shape[1]))
    for i in model.X:         # estimate pairwise probabilities
        for j in model.X:
            if j<=i: continue   # estimate (empirical) mutual information:
            wts[i,j] = (phat[i,j] * (phat[i,j]/phat[i]/phat[j]).log() ).sum()

    np.set_printoptions(precision=4, suppress=True)
    print(np.round(wts,3))
    
    return wts
```

```python
model, Vars = initializeGraphModel()
train, valid = loadData()
```

```python
phat = estimateProb(model, train)
```

```python
wts = computeEmpirical(model, phat)
```

```python
G = nx.Graph()

pos = {Vars[0]: (0,3), 
       Vars[1]:(1,3), 
       Vars[2]:(2,3), 
       Vars[3]:(3,3), 
       Vars[4]:(4,3), 
       Vars[5]:(1,2), 
       Vars[6]:(2,2), 
       Vars[7]:(3,2), 
       Vars[8]:(1,1), 
       Vars[9]:(2,1),
       Vars[10]:(2,0)
      }

labelDict = {Vars[0]: "jgKilled", 
             Vars[1]: "turrets",
             Vars[2]: "csDiff",
             Vars[3]: "wardDiff",
             Vars[4]: "eliteMon",
             Vars[5]: "goldDiff",
             Vars[6]: "expDiff",
             Vars[7]: "avgKDA",
             Vars[8]: "laneDom",
             Vars[9]: "teamplay",
             Vars[10]: "win"}

labels = {}

for Xi in Vars:
    G.add_node(Xi.label)

for i in range(train.shape[1]):
    for j in range(i, train.shape[1]):
        G.add_edge(i,j, weight=wts[i,j])
        labels[(i,j)] = np.round(wts[i,j], 4)
```

```python
nx.draw(G, pos, labels=labelDict, with_labels=True)
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)
```

```python
mst = nx.maximum_spanning_tree(G)
nx.draw(mst, pos, labels=labelDict, with_labels=True)
```
